```python
import kagglehub
import os
from pathlib import Path
```

```python
cwd = Path.cwd()
ROOT = cwd.parent
DATA_DIR = ROOT / "data"
DATA_DIR.mkdir(exist_ok=True)
```

```python
def download_dataset():
    # Download latest version
    
    kagglehub.dataset_download('arjuntejaswi/plant-village', output_dir=DATA_DIR)
```

```python
download_dataset()
```

```python
SOURCE_DIR = Path("PlantVillage")
SOURCE_DIR
```

```python
import shutil
from sklearn.model_selection import train_test_split

SOURCE_DIR = DATA_DIR / "PlantVillage"
OUTPUT_DIR = DATA_DIR / "PlantVillage_split"

train_ratio = 0.70
val_ratio = 0.15
test_ratio = 0.15

image_exts = {".jpg", ".jpeg", ".png", ".webp"}

for class_dir in SOURCE_DIR.iterdir():

    if not class_dir.is_dir():
        continue

    class_name = class_dir.name

    images = [
        p for p in class_dir.iterdir()
        if p.suffix.lower() in image_exts
    ]

    train_imgs, temp_imgs = train_test_split(
        images,
        test_size=(val_ratio + test_ratio),
        random_state=42,
        shuffle=True
    )

    val_imgs, test_imgs = train_test_split(
        temp_imgs,
        test_size=test_ratio / (val_ratio + test_ratio),
        random_state=42,
        shuffle=True
    )

    for split_name, split_imgs in {
        "train": train_imgs,
        "val": val_imgs,
        "test": test_imgs
    }.items():

        split_dir = OUTPUT_DIR / split_name / class_name
        split_dir.mkdir(parents=True, exist_ok=True)

        for img_path in split_imgs:
            shutil.copy2(img_path, split_dir / img_path.name)

print("Dataset split complete.")
```
